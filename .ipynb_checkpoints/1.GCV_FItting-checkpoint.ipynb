## Getting the data

```python
Data = pickle.load( open( "ts1.p", "rb" ) )
```

## Plotting

```python
#figure(figsize=(20,8))
f1 = figure(figsize=(15,10))
ax = subplot2grid((1,1),(0,0))
####### Scatter plot for the smaller time series
p = 4;q=2
scatter(Data[:,0],Data[:,1],color = 'r',s = 100,label = 'Data')
[n,lamb,sigmasq] = full_search_nk(Data,p,q)
c = n+p
U = Kno_pspline_opt(Data,p,n)
B = Basis_Pspline(n,p,U,Data[:,0])
P = Penalty_p(q,c)
theta = np.linalg.solve(B.T.dot(B) + lamb*P, B.T.dot(Data[:,1].reshape(-1,1)))
### Getting mean of the prediction
num = 200
xpred = linspace(Data[0,0],Data[-1,0],num)
Bpred = Basis_Pspline(n,p,U,xpred)
ypred1 = Bpred.dot(theta)
std_t1,std_n1 = Var_bounds(Data,Bpred,B,theta,P,lamb)


ax.plot(xpred,ypred1,linewidth=3,color = 'g',label = 'Mean Prediction')
ax.set_title('(a)',size = 25)
ax.tick_params(axis='x', labelsize=19)
ax.tick_params(axis='y', labelsize=19)
ax.set_xlabel('Time',size=25)
ax.set_ylabel('thickness change (m)',size = 25)
ax.fill_between(xpred.flatten(),ypred1.flatten()-std_t1,ypred1.flatten()+std_t1, alpha = 0.2,color = 'k',label = '95% t-interval')
ax.legend(fontsize=20)
ax.grid(True)
show()
```

```python
def plotting_GCV(Data,p,q,n,lamb):
    c = n+p
    U = Kno_pspline_opt(Data,p,n)
    B = Basis_Pspline(n,p,U,Data[:,0])
    
    P = lamb*Penalty_p(q,c)
    H = B.dot(inv(B.T.dot(B)+P)).dot(B.T)
    y_cap = H.dot(Data[:,1].reshape(-1,1))
    d = sum(diag(H))/n
    t = 0
    for i in range(n):
        t = t+ ((Data[i,1] - y_cap[i])/(1-d))**2

    #obj = (1/n)*t
    obj = t
    return obj
```

```python
gcv = []


font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

f1 = figure(figsize=(7,5))
l = list(np.linspace(0.001,0.5,200))
#l.insert(0,0.000001)
[n,lamb,sigmasq] = full_search_nk(Data,p,q)
for lamb1 in l:
    gcv.append(plotting_GCV(Data,p,q,n,lamb1))

#f1 = figure(figsize=(15,10))
ax = subplot2grid((1,1),(0,0))
ax.plot(l,gcv,color = 'k')
ax.scatter(0.001,plotting_GCV(Data,p,q,n,0.001),color = 'k',marker = '*',s = 200)
ax.scatter(0.01,plotting_GCV(Data,p,q,n,0.01),color = 'k',marker = '*',s = 200)
ax.scatter(0.05,plotting_GCV(Data,p,q,n,0.05),color = 'k',marker = '*',s = 200)
ax.scatter(0.5,plotting_GCV(Data,p,q,n,0.5),color = 'k',marker = '*',s = 200)

text(0.03, 290, r'$\lambda = 0.001$', fontdict=font,size = 20)
text(0.04, 230, r'$\lambda = 0.01$', fontdict=font,size = 20)
text(0.2, 200, r'$\lambda = 0.05$', fontdict=font,size = 20)
text(0.43, 240, r'$\lambda = 0.5$', fontdict=font,size = 20)

ax.tick_params(axis='x', labelsize=25)
ax.tick_params(axis='y', labelsize=25)
ax.set_xlabel('$\lambda$',size=25)
ax.set_ylabel('$GCV(\lambda)$',size = 25)
#ax.legend(fontsize=25)
#grid(True)  
f1.savefig("fig_temp2.pdf", bbox_inches='tight')
show()
```
